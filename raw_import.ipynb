```python
import numpy as np
from PIL import Image
from pathlib import Path
path1 = '/Users/danijelmaricic/Downloads/'
file_name = 'Image__2024-05-31__14-08-37.raw'
file_name_only = Path(file_name).stem
file_name_raw = file_name_only + '-OUT.RAW'
```

```python
def read_write_raw(): 
    with open(path1 + file_name, 'r') as infile:
        data = np.fromfile(infile, dtype=np.uint8) 
        pixels, = data.shape 
        height = int(2160) 
        width = int(pixels / height) 
        data = np.reshape(data, (height, width)) 
        im = Image.fromarray(data) 
        im.save('template1.png') 
        data.astype(np.uint16).tofile(path1 + file_name_raw)
```

```python
read_write_raw()
```
